```r
#Urn
```

```r
x<-c(3,5,10,15,20,25,28,34)
y<-c(4,10,17,16,10,5,3.5,7)
```

```r
plot(x,y,col="blue", main="Urn Shape",xlab="x coordinate (in cm)", ylab="y coordinate (in cm)")
```

```r
relation3<-lm(y~I(x)+I(x^2)+I(x^3))
print(relation3)
summary(relation3)
```

```r
x_seq <- seq(from=3,to=34, by=0.01)
new_relation3<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation3[i]<-predict(relation3, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation3),xlab="x",ylab="y")
```

```r
RSE_Percent = 0.3112/mean(y) * 100
```

```r
RSE_Percent
```

```r
confint(relation3, level=0.95)
```

```r
plot(fitted(relation3), residuals(relation3))
```

```r
#Soda Bottle
```

```r
x<-c(1,2,3,4,5,6,7,8,9,9.875)
y<-c(0.938,1.375,1.156,1.250,1.313,1.313,1.281,0.813,0.5,0.625)
```

```r
plot(x,y,col="blue", main="Soda Bottle Shape",xlab="x coordinate (in cm)", ylab="y coordinate (in cm)")
```

```r
relation4<-lm(y~I(x)+I(x^2)+I(x^3)+I(x^4))
print(relation4)
summary(relation4)
```

```r
RSE_Percent = 0.1793/mean(y) * 100
```

```r
RSE_Percent
```

```r
confint(relation4, level=0.95)
```

```r
plot(fitted(relation4), residuals(relation4))
```
